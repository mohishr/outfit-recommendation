```python
def filter_outfits_by_category(json_data, allowed_categories):
    """
    Keep outfits that contain at least one item with categoryid in allowed_categories.
    """
    filtered = []

    for outfit in json_data:
        items = outfit.get("items", [])
        for item in items:
            if item.get("categoryid") in allowed_categories:
                filtered.append(outfit)
                break   # outfit accepted, no need to check more items

    return filtered[:500]
```

```python
import json

# Load your JSON dataset
with open("C:\\Users\\shrey\\source\\repos\\outfit-gnn\\data\\polyvore\\test_no_dup.json", "r") as f:
    data = json.load(f)

allowed_categories = {
    15, 17, 18, 19,
    21, 23, 24, 25, 26, 27, 28, 29, 30
}

filtered_data = filter_outfits_by_category(data, allowed_categories)

print("Original:", len(data))
print("Filtered:", len(filtered_data))

# Save filtered result if needed
with open("filtered_data.json", "w") as f:
    json.dump(filtered_data, f, indent=4)
```

```python
import os
import shutil
from pathlib import Path

def copy_outfit_folders(filtered_data, src_root, dst_root):
    """
    Copy folders matching set_id from src_root → dst_root.

    filtered_data : list of outfit dicts
    src_root      : directory containing folders with set_id names
    dst_root      : destination directory
    """

    src_root = Path(src_root)
    dst_root = Path(dst_root)

    dst_root.mkdir(parents=True, exist_ok=True)

    # collect unique set_ids
    set_ids = {str(outfit["set_id"]) for outfit in filtered_data}

    print(f"Found {len(set_ids)} set_ids to copy.")

    for set_id in set_ids:
        src = src_root / set_id
        dst = dst_root / set_id

        if src.exists() and src.is_dir():
            print(f"Copying {set_id} → {dst}")
            shutil.copytree(src, dst, dirs_exist_ok=True)
        else:
            print(f"WARNING: Folder not found → {src}")
```
